```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
```

```python
num_imputer = SimpleImputer(strategy="median")
df[num_cols] = num_imputer.fit_transform(df[num_cols])
```

```python
cat_imputer = SimpleImputer(strategy="most_frequent")
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
```

```python
num_cols
```

```python
date_cols = ['Policy_Start_Date', 'Policy_End_Date', 'Last_Claim_Date', 'Last_Payment_Date']
df[date_cols] = df[date_cols].apply(pd.to_datetime, errors='coerce')
anchor_date = df[date_cols].max().max()
```

```python
df['Policy_Age'] = (anchor_date - df['Policy_Start_Date']).dt.days
df['Policy_Duration'] = (df['Policy_End_Date'] - df['Policy_Start_Date']).dt.days
df['Time_Since_Last_Payment'] = (anchor_date - df['Last_Payment_Date']).dt.days
df['Time_Since_Last_Claim'] = (anchor_date - df['Last_Claim_Date']).dt.days
df['Has_Filed_Claim'] = df['Last_Claim_Date'].notna().astype(int)
df['Time_Since_Last_Claim'].fillna(df['Time_Since_Last_Claim'].median(), inplace=True)
```

```python
df.drop(columns=date_cols, inplace=True)
df.drop(columns='Customer_ID', inplace=True)
```

```python
# num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
# cat_cols = df.select_dtypes(include=['object']).columns.tolist()
cat_cols = ['Gender', 'Marital_Status', 'Region', 'Education_Level', 'Employment_Status', 
            'Policy_Type', 'Mode_of_Communication', 'Payment_Mode', 'Vehicle_Type']
num_cols = ['Age', 'Policy_Tenure', 'Policy_Renewal_Count', 'Total_Claims_Filed', 'Claim_Amount_Total', 
            'Missed_Payments', 'Family_Members_Insured', 'Policy_Cancellation_History', 'Customer_Complaints', 
            'Complaint_Resolution_Time', 'Customer_Support_Calls', 'Vehicle_Age', 'Premium_Amount', 
            'Policy_Age', 'Policy_Duration', 'Time_Since_Last_Payment', 'Time_Since_Last_Claim']
```

```python
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded = pd.DataFrame(encoder.fit_transform(df[cat_cols]), columns=encoder.get_feature_names_out(cat_cols))
```

```python
df = df.drop(columns=cat_cols).reset_index(drop=True)
df = pd.concat([df, encoded], axis=1)
```

```python
scaler = StandardScaler()
scaled = pd.DataFrame(scaler.fit_transform(df[num_cols]), columns=num_cols)
```

```python
df_final = pd.concat([scaled, encoded], axis=1)
df_final['Churn'] = df['Churn'].values
```

```python
joblib.dump(encoder, r"models\encoder.pkl")
joblib.dump(scaler, r"models\scaler.pkl")
joblib.dump(anchor_date, r"models\anchor_date.pkl")
```

```python
df_final.to_csv(r"models\processed_data.csv", index=False)
```
